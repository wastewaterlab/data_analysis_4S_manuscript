## Limit of detection summary

```python
t='N1'
df= raw_outliers_flagged_df
df=df[df.Target==t].copy()
df=df[df.Task=='Standard'].copy()
df2=df[df.grubbs_test==False].groupby("Quantity")["grubbs_test"].agg('count').reset_index()
df_t=df[df.grubbs_test==True].copy()
df_t=df_t[df_t.Quantity!=0]
df_t["log10_conc"]=np.log10(df_t.Quantity)
out=df_t.groupby(["log10_conc","Quantity"]).agg( p_id=('plate_id','min'),
                                    positives=('Cq','count'),
                                    total=('Sample', 'count')).reset_index()
out['fr_pos']=out.positives/out.total
out=out.merge(df2,how="left", on="Quantity")
out.grubbs_test=out.grubbs_test.fillna(0)#when all pass grubbs
N1=out[["Quantity","fr_pos","total","grubbs_test"]].copy()



t='PMMoV'
df= raw_outliers_flagged_df
df=df[df.Target==t].copy()
df=df[df.Task=='Standard'].copy()
df2=df[df.grubbs_test==False].groupby("Quantity")["grubbs_test"].agg('count').reset_index()
df_t=df[df.grubbs_test==True].copy()
df_t=df_t[df_t.Quantity!=0]
df_t["log10_conc"]=np.log10(df_t.Quantity)
out=df_t.groupby(["log10_conc","Quantity"]).agg( p_id=('plate_id','min'),
                                    positives=('Cq','count'),
                                    total=('Sample', 'count')).reset_index()
out['fr_pos']=out.positives/out.total
out=out.merge(df2,how="outer", on="Quantity")
out.grubbs_test=out.grubbs_test.fillna(0)#when all pass grubbs
P=out[["Quantity","fr_pos","total","grubbs_test"]].copy()

t='bCoV'
df= raw_outliers_flagged_df
df=df[df.Target==t].copy()
df=df[df.plate_id==32].copy()
df=df[df.Task=='Standard'].copy()
df2=df[df.grubbs_test==False].groupby("Quantity")["grubbs_test"].agg('count').reset_index()
df_t=df[df.grubbs_test==True].copy()
df_t=df_t[df_t.Quantity!=0]
df_t["log10_conc"]=np.log10(df_t.Quantity)
out=df_t.groupby(["log10_conc","Quantity"]).agg( p_id=('plate_id','min'),
                                    positives=('Cq','count'),
                                    total=('Sample', 'count')).reset_index()
out['fr_pos']=out.positives/out.total
out=out.merge(df2,how="left", on="Quantity")
out.grubbs_test=out.grubbs_test.fillna(0)#when all pass grubbs
bCoV=out[["Quantity","fr_pos","total","grubbs_test"]].copy()

t='18S'
df= raw_outliers_flagged_df
df=df[df.Task=='Standard'].copy()
df=df[df.Target==t].copy()
df2=df[df.grubbs_test==False].groupby("Quantity")["grubbs_test"].agg('count').reset_index()

df_t=df[df.grubbs_test==True].copy()
df_t=df_t[df_t.Quantity!=0]
df_t["log10_conc"]=np.log10(df_t.Quantity)
out=df_t.groupby(["log10_conc","Quantity"]).agg( p_id=('plate_id','min'),
                                    positives=('Cq','count'),
                                    total=('Sample', 'count')).reset_index()
out['fr_pos']=out.positives/out.total
out=out.merge(df2,how="left", on="Quantity")
out.grubbs_test=out.grubbs_test.fillna(0)#when all pass grubbs
S18=out[["Quantity","fr_pos","total","grubbs_test"]].copy()

N1["assay"]="N1"
P["assay"]="PMMoV"
bCoV["assay"]="bCoV"
S18["assay"]="18S"

table_S5=pd.concat([N1,P,bCoV,S18])
table_S5["total_reps"]=table_S5["total"]+table_S5["grubbs_test"]
table_S5=table_S5[["assay", "Quantity", "fr_pos","total", "grubbs_test", "total_reps"]]
table_S5
```

```python
a=inhib_data.copy()
a[a.Target!='bCoV']
subset_code='wash_buffer'
category_order=['1 mL / 2 mL', '3 mL / 6 mL', '5 mL / 10 mL', '7 mL / 14 mL', '10 mL / 20 mL', '25 mL / 50 mL']

a=a[a.experiment== subset_code].copy()
a.sample_experiment_detail=a.sample_experiment_detail.astype('category')
a.sample_experiment_detail=a.sample_experiment_detail.cat.reorder_categories(category_order)
a['Target']=a['Target'].str.replace('N1','SARS-CoV-2 N1')
a['Target']=a['Target'].astype('category')
a['Target']=a['Target'].cat.reorder_categories(['SARS-CoV-2 N1','PMMoV','bCoV'])

a=a[["Sample", "Target","dilution", "sample_experiment_detail", 'exp_dCt_perfect', 'dCt']].copy()
a_melt=a.melt(id_vars=["Sample", "Target","dilution", "sample_experiment_detail"], value_vars=['exp_dCt_perfect', 'dCt'],var_name='' )
a_melt['ymin_p']=a_melt.value -1 
a_melt['ymax_p']=a_melt.value +1
a_melt.loc[a_melt[""]=="dCt", 'ymin_p']=np.nan
a_melt.loc[a_melt[""]=="dCt", 'ymax_p']=np.nan
a_melt.loc[a_melt[""]=="dCt",""]= "experimental dCt"
a_melt.loc[a_melt[""]=="exp_dCt_perfect",""]= "theoretical dCt"

fig6s=(ggplot(a_melt) + 
geom_ribbon(aes(x='dilution',  fill="", ymin='ymin_p',ymax='ymax_p'), alpha=0.3)+
geom_point(aes(x='dilution', y='value', color=""), size=5)+
geom_line(aes(x='dilution', y='value', color=""))+
scale_color_manual([ADA_c,case2_c])+
scale_fill_manual([ADA_c,case2_c])+
facet_grid('Target ~ sample_experiment_detail')+
ylab('dCt')+
theme_classic()+ theme(legend_position='top',axis_title=element_text(size=20), strip_text=element_text(size=12),  legend_text=element_text(size=12),axis_text_x=element_text(rotation=45, hjust=1, size=12), axis_text_y=element_text(size=12),figure_size=(18,9)) )

fig6s
```
