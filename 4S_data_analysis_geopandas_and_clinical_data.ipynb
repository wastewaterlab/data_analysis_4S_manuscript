## Limit of detection summary

```python
t='N1'
df= raw_outliers_flagged_df
df=df[df.Target==t].copy()
df=df[df.Task=='Standard'].copy()
df2=df[df.grubbs_test==False].groupby("Quantity")["grubbs_test"].agg('count').reset_index()
df_t=df[df.grubbs_test==True].copy()
df_t=df_t[df_t.Quantity!=0]
df_t["log10_conc"]=np.log10(df_t.Quantity)
out=df_t.groupby(["log10_conc","Quantity"]).agg( p_id=('plate_id','min'),
                                    positives=('Cq','count'),
                                    total=('Sample', 'count')).reset_index()
out['fr_pos']=out.positives/out.total
out=out.merge(df2,how="left", on="Quantity")
out.grubbs_test=out.grubbs_test.fillna(0)#when all pass grubbs
N1=out[["Quantity","fr_pos","total","grubbs_test"]].copy()



t='PMMoV'
df= raw_outliers_flagged_df
df=df[df.Target==t].copy()
df=df[df.Task=='Standard'].copy()
df2=df[df.grubbs_test==False].groupby("Quantity")["grubbs_test"].agg('count').reset_index()
df_t=df[df.grubbs_test==True].copy()
df_t=df_t[df_t.Quantity!=0]
df_t["log10_conc"]=np.log10(df_t.Quantity)
out=df_t.groupby(["log10_conc","Quantity"]).agg( p_id=('plate_id','min'),
                                    positives=('Cq','count'),
                                    total=('Sample', 'count')).reset_index()
out['fr_pos']=out.positives/out.total
out=out.merge(df2,how="outer", on="Quantity")
out.grubbs_test=out.grubbs_test.fillna(0)#when all pass grubbs
P=out[["Quantity","fr_pos","total","grubbs_test"]].copy()

t='bCoV'
df= raw_outliers_flagged_df
df=df[df.Target==t].copy()
df=df[df.plate_id==32].copy()
df=df[df.Task=='Standard'].copy()
df2=df[df.grubbs_test==False].groupby("Quantity")["grubbs_test"].agg('count').reset_index()
df_t=df[df.grubbs_test==True].copy()
df_t=df_t[df_t.Quantity!=0]
df_t["log10_conc"]=np.log10(df_t.Quantity)
out=df_t.groupby(["log10_conc","Quantity"]).agg( p_id=('plate_id','min'),
                                    positives=('Cq','count'),
                                    total=('Sample', 'count')).reset_index()
out['fr_pos']=out.positives/out.total
out=out.merge(df2,how="left", on="Quantity")
out.grubbs_test=out.grubbs_test.fillna(0)#when all pass grubbs
bCoV=out[["Quantity","fr_pos","total","grubbs_test"]].copy()

t='18S'
df= raw_outliers_flagged_df
df=df[df.Task=='Standard'].copy()
df=df[df.Target==t].copy()
df2=df[df.grubbs_test==False].groupby("Quantity")["grubbs_test"].agg('count').reset_index()

df_t=df[df.grubbs_test==True].copy()
df_t=df_t[df_t.Quantity!=0]
df_t["log10_conc"]=np.log10(df_t.Quantity)
out=df_t.groupby(["log10_conc","Quantity"]).agg( p_id=('plate_id','min'),
                                    positives=('Cq','count'),
                                    total=('Sample', 'count')).reset_index()
out['fr_pos']=out.positives/out.total
out=out.merge(df2,how="left", on="Quantity")
out.grubbs_test=out.grubbs_test.fillna(0)#when all pass grubbs
S18=out[["Quantity","fr_pos","total","grubbs_test"]].copy()

N1["assay"]="N1"
P["assay"]="PMMoV"
bCoV["assay"]="bCoV"
S18["assay"]="18S"

table_S5=pd.concat([N1,P,bCoV,S18])
table_S5["total_reps"]=table_S5["total"]+table_S5["grubbs_test"]
table_S5=table_S5[["assay", "Quantity", "fr_pos","total", "grubbs_test", "total_reps"]]
table_S5
```

```python
shape_EBMUD_zipC= gpd.sjoin(shapeCA_zipC,shape_EBMUD_service,how='right',op='intersects').reset_index()

shape_A= shape_EBMUD_service[shape_EBMUD_service.Interceptor=='A'].dissolve(by='Interceptor').reset_index()
shape_N= shape_EBMUD_service[shape_EBMUD_service.Interceptor=='N'].dissolve(by='Interceptor').reset_index()
shape_S= shape_EBMUD_service[shape_EBMUD_service.Interceptor=='S'].dissolve(by='Interceptor').reset_index()
```

### Weight Case data by the overlap

```python
#overlay the areas, calculate the area of the overlap and the total area for each zipcode
inter_A = gpd.overlay(shape_A, shapeCA_zipC)
inter_A['inter_A'] = inter_A.area #intersection area
inter_A= pd.DataFrame(inter_A[['ZCTA','inter_A']].copy())

inter_N = gpd.overlay(shape_N,shapeCA_zipC)
inter_N['inter_N'] = inter_N.area #intersection area
inter_N= pd.DataFrame(inter_N[['ZCTA','inter_N']].copy())

inter_S = gpd.overlay(shape_S,shapeCA_zipC)
inter_S['inter_S'] = inter_S.area #intersection area
inter_S= pd.DataFrame(inter_S[['ZCTA','inter_S']].copy())

inter_all=inter_A.append(inter_N)
inter_all=inter_all.append(inter_S)
shapeCA_zipC['tot_area']=shapeCA_zipC.area
shapeCA_zipC=shapeCA_zipC.merge(inter_all,how='left',on='ZCTA')


# #calculate the f
shapeCA_zipC['fr_A']=shapeCA_zipC['inter_A']/shapeCA_zipC['tot_area']
shapeCA_zipC['fr_N']=shapeCA_zipC['inter_N']/shapeCA_zipC['tot_area']
shapeCA_zipC['fr_S']=shapeCA_zipC['inter_S']/shapeCA_zipC['tot_area']

#multiply by A fraction
shapeCA_zipC_A=shapeCA_zipC.copy()
shapeCA_zipC_A[shapeCA_zipC_A.select_dtypes([np.float64]).columns]=shapeCA_zipC_A[shapeCA_zipC_A.select_dtypes([np.float64]).columns].multiply(shapeCA_zipC_A['fr_A'], axis='index')
shapeCA_zipC_A=shapeCA_zipC_A.drop(["ZCTA", "inter_A","inter_N","inter_S","fr_A","fr_N","fr_S","index"],axis=1)
s_A=pd.DataFrame(shapeCA_zipC_A.sum(axis=0))

# #multiply by N fraction
shapeCA_zipC_N=shapeCA_zipC.copy()
shapeCA_zipC_N[shapeCA_zipC_N.select_dtypes([np.float64]).columns]=shapeCA_zipC_N[shapeCA_zipC_N.select_dtypes([np.float64]).columns].multiply(shapeCA_zipC_N['fr_N'], axis='index')
shapeCA_zipC_N=shapeCA_zipC_N.drop(["ZCTA", "inter_A","inter_N","inter_S","fr_A","fr_N","fr_S","index"],axis=1)
s_N=pd.DataFrame(shapeCA_zipC_N.sum(axis=0))


# #multiply by S fraction
shapeCA_zipC_S=shapeCA_zipC.copy()
shapeCA_zipC_S[shapeCA_zipC_S.select_dtypes([np.float64]).columns]=shapeCA_zipC_S[shapeCA_zipC_S.select_dtypes([np.float64]).columns].multiply(shapeCA_zipC_S['fr_S'], axis='index')
shapeCA_zipC_S=shapeCA_zipC_S.drop(["ZCTA", "inter_A","inter_N","inter_S","fr_A","fr_N","fr_S","index"],axis=1)
s_S=pd.DataFrame(shapeCA_zipC_S.sum(axis=0))
```
